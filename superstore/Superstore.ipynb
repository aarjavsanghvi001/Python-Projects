```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('A:/Datasets/Superstore.csv')
data.head(10)
```

```python
data.isnull().sum()
```

```python
data = data.rename(columns={'Order ID':'OrderID'})
data
```

```python
data=data.drop(['OrderID'],axis=1)
data.head(2)
```

```python
data['State'].fillna(str(data['State'].mode().values[0]),inplace=True)
```

```python
data['Postal Code'].fillna(data['Postal Code'].median(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['State'].unique()
```

```python
data=data.drop(['Country/Region'],axis=1)
data.head(2)
```
